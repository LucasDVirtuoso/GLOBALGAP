## 2. Carregamento de dados, limpeza e tratamento

```python
# Carrega os dados de exportação do Chile
ch_2018 = pd.read_excel('/content/drive/MyDrive/Pesquisa - Talles/Padrões de normas vigentes no comércio internacional: caracterização e análise do efeito da norma Globalgap sobre as exportações de frutas no Brasil, Chile e Argentina./Dados/Exportações/dados_brutos/Chile/Exportações/expchile_2018.xlsx')
ch_2019 = pd.read_excel('/content/drive/MyDrive/Pesquisa - Talles/Padrões de normas vigentes no comércio internacional: caracterização e análise do efeito da norma Globalgap sobre as exportações de frutas no Brasil, Chile e Argentina./Dados/Exportações/dados_brutos/Chile/Exportações/expchile_2019.xlsx')
ch_2020 = pd.read_excel('/content/drive/MyDrive/Pesquisa - Talles/Padrões de normas vigentes no comércio internacional: caracterização e análise do efeito da norma Globalgap sobre as exportações de frutas no Brasil, Chile e Argentina./Dados/Exportações/dados_brutos/Chile/Exportações/expchile_2020.xlsx')
ch_2021 = pd.read_excel('/content/drive/MyDrive/Pesquisa - Talles/Padrões de normas vigentes no comércio internacional: caracterização e análise do efeito da norma Globalgap sobre as exportações de frutas no Brasil, Chile e Argentina./Dados/Exportações/dados_brutos/Chile/Exportações/expchile_2021.xlsx')
ch_2022 = pd.read_excel('/content/drive/MyDrive/Pesquisa - Talles/Padrões de normas vigentes no comércio internacional: caracterização e análise do efeito da norma Globalgap sobre as exportações de frutas no Brasil, Chile e Argentina./Dados/Exportações/dados_brutos/Chile/Exportações/expchile_2022.xlsx')
```

```python
# Cria lista com os dados de cada ano
chile = [ch_2018, ch_2019, ch_2020, ch_2021, ch_2022]
```

```python
# Renomeia as colunas do DataFrame
for i in chile:
  i.columns = ['uf', 'ncm', 'produto', 'unidade', 'peso_kg', 'fob_valor']
```

```python
# Remove coluna ['produto']
for i in chile:
  i.drop(i.iloc[:, [2]], axis=1, inplace=True)
```

```python
# Cria dicionário para corrigir valores
# Corrige códigos para os nomes das frutas

novos_valores = {
    8044000: 'abacate',
    8045000: 'manga',
    8055010: 'limao',
    8061000: 'uva',
    8071100: 'melancia',
    8071900: 'melao',
    8072000: 'mamao',
    8081000: 'maca',
    8055090: 'limas_limones_secosoufrescos',
    8055020: 'lima_agria'
}
```

```python
# Substitui os códigos pelos nomes das frutas
for i in chile:
  i['ncm'] = i['ncm'].replace(novos_valores)
```
