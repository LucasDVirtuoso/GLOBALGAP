- Altera nome das Colunas para facilitar manipulação

```python
# Cria lista pra a Argentina
argentina = [ar_2018, ar_2019, ar_2020, ar_2021, ar_2022]
```

```python
# Renomeia colunas
for i in argentina:
  i.columns = [['ano', 'uf', 'ncm', 'rubro', 'pais', 'fob_dolar', 'peso_kg']]
```

```python
# Remove colunas desnecessárias no momento
for i in argentina:
  i.drop(i.iloc[:, [0, 3, 4]], axis=1, inplace=True)
```

- Verificação de valores nulos
- Verificação de tipos de dados que estão estruturados no DataFrame

```python
# Loop para verificar valores nulos
yr = 2018 # Marca o ano analizado
for dado in argentina:
  variavel = dado.isnull().sum()
  print(yr)
  print(variavel)
  yr += 1 # Aumenta o ano
```

- Não foram apresentados valores nulos na base, considerando apenas as frutas da amostra disponíveis.

- Substituir os valores do código ncm para o nome da fruta correspondente.
- Filtrar as frutas: Deixar apenas as que correspondem à amostra.

```python
# Cria dicionário com novos valores a serem substituidos
novos_valores = {
    8044000.0: 'abacate',
    8055000.0: 'limao_e_lima',
    8061000.0: 'uva',
    8081000.0: 'maca'
}

# Substitui valores antigos pelos novos
for i in argentina:
  i['ncm'] = i['ncm'].replace(novos_valores)
```

```python
# Cria dicionário com novos valores para substituir as siglas das provincias pelo nome completo
nome_provincias = {
    'TN': 'tucuman',
    'BA': 'buenos aires',
    'CA': 'catamarca',
    'ER': 'entre rios',
    'FO': 'formosa',
    'JU': 'jujuy',
    'MI': 'misiones',
    'RR': 'corrientes',
    'SA': 'salta',
    'ZA': 'mendoza',
    'NE': 'neuquen',
    'RN': 'rio negro',
    'IN': 'indeterminado',
    'NN': 'san juan'
}

# Substitui as siglas pelos nomes em todos os DataFrames correspondentes à Argentina
for i in argentina:
  i['uf'] = i['uf'].replace(nome_provincias)
```

```python
# Contando valores por provincia (uf)
yr = 2018 # Marca o ano inicial da contagem
for i in argentina:
  provincia = i['uf'].value_counts()
  print(yr)
  print(provincia)
  yr += 1
```
